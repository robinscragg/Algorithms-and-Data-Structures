````python
class Node:
  def __init__(self, value, value2, next_node=None, prev_node=None):
    self.value = value
    self.next_node = next_node
    self.prev_node = prev_node
    self.value2 = value2
    
  def get_value(self):
    return self.value

  def get_value2(self):
    return self.value2
  
  def get_next_node(self):
    return self.next_node
  
  def set_next_node(self, next_node):
    self.next_node = next_node
    
  def get_prev_node(self):
    return self.prev_node
  
  def set_prev_node(self, prev_node):
    self.prev_node = prev_node
    
    
class LinkedList:
  def __init__(self, value=None, value2=None):
    self.head_node = Node(value, value2)
    self.tail_node = Node(value, value2)
  
  def get_head_node(self):
     return self.head_node  

  def get_tail_node(self):
     return self.tail_node

  def insert_beginning(self, new_value, new_value2):
    new_head = Node(new_value, new_value2)
    current_head = self.head_node

    if current_head != None:
      current_head.set_prev_node(new_head)
      new_head.set_next_node(current_head)

    self.head_node = new_head

    if self.tail_node == None:
      self.tail_node = new_head 
    
  def insert_end(self, new_value, new_value2):
    new_tail = Node(new_value, new_value2)
    current_tail = self.tail_node
    
    if current_tail != None:
      current_tail.set_next_node(new_tail)
      new_tail.set_prev_node(current_tail)

    self.tail_node = new_tail

    if self.head_node == None:
      self.head_node = new_tail
    
  def stringify_list(self):
    string_list = ""
    current_node = self.get_head_node()
    while current_node:
      if current_node.get_value() != None:
        string_list += str(current_node.get_value()) + " " + str(current_node.get_value2()) + "\n"
      current_node = current_node.get_next_node()      
    return string_list

  def find_node(self, value_to_find):
    current_node = self.get_head_node()
    found = False
    while current_node:
        if current_node.get_value() == value_to_find or current_node.get_value2() == value_to_find:
            found = True
            current_node = None       
        else:
          current_node = current_node.get_next_node()
    if found == True:
        print("Found the value: "+ str(value_to_find))
    else:
        print("Cannot find the value: " + str(value_to_find))  
        
  def remove_head(self):
    removed_head = self.head_node

    if removed_head == None:
      return None

    self.head_node = removed_head.get_next_node()

    if self.head_node != None:
      self.head_node.set_prev_node(None)

    if removed_head == self.tail_node:
      self.remove_tail()
    
    return removed_head.get_value()


  def remove_tail(self):
    removed_tail = self.tail_node

    if removed_tail == None:
      return None

    self.tail_node = removed_tail.get_prev_node()

    if self.tail_node != None:
      self.tail_node.set_next_node(None)

    if removed_tail == self.head_node:
      self.remove_head()
    
    return removed_tail.get_value()
      
        
  def remove_by_value(self, value_to_remove):
    if type(value_to_remove) == int:
        value_to_remove = str(value_to_remove)
    print("Removing: " + value_to_remove)
    node_to_remove = None
    current_node = self.head_node

    while current_node != None:
      if current_node.get_value() == value_to_remove:
        node_to_remove = current_node
        break

      current_node = current_node.get_next_node()

    if node_to_remove == None:
      return None

    if node_to_remove == self.head_node:
      self.remove_head()
    elif node_to_remove == self.tail_node:
      self.remove_tail()
    else:
      next_node = node_to_remove.get_next_node()
      prev_node = node_to_remove.get_prev_node()
      next_node.set_prev_node(prev_node)
      prev_node.set_next_node(next_node)

    return node_to_remove
   

linkedList = LinkedList("apple", 5)
linkedList.insert_beginning("banana", 10)
linkedList.insert_beginning("pear", 8)
linkedList.insert_beginning("orange", 7)
linkedList.insert_beginning("mango", 4)

print(linkedList.stringify_list())

linkedList.find_node("mango")
linkedList.find_node("kiwi")

print(linkedList.stringify_list())
linkedList.remove_by_value("mango")
print(linkedList.stringify_list())

print(linkedList.stringify_list())
linkedList.remove_by_value("pear")
print(linkedList.stringify_list())

linkedList.find_node(5)

linkedList.insert_end("guava", 1)
linkedList.insert_end("lemon", 3)
print(linkedList.stringify_list())
````

````python
import random
import time

def randomArray(array_size):
    array = [random.randint(0,9) for i in range(array_size)]
    return array

def swap(arr, index_1, index_2):
  temp = arr[index_1]
  arr[index_1] = arr[index_2]
  arr[index_2] = temp

def bubble_sort_unoptimized(arr):
  iteration_count = 0
  for el in arr:
    for index in range(len(arr) - 1):
      iteration_count += 1
      if arr[index] > arr[index + 1]:
        swap(arr, index, index + 1)
  return arr

def bubble_sort_optimized(arr):
  iteration_count = 0
  for i in range(len(arr)):
    # iterate through unplaced elements
    for idx in range(len(arr) - i - 1):
      iteration_count += 1
      if arr[idx] > arr[idx + 1]:
        # replacement for swap function
        arr[idx], arr[idx + 1] = arr[idx + 1], arr[idx]    
  return arr
    
def linear_search(search_list, target_value):
  for idx in range(len(search_list)):
    if search_list[idx] == target_value:
      return idx
  raise ValueError("{0} not in list".format(target_value))
    
def binary_search(sorted_list, target):
  left_pointer = 0
  right_pointer = len(sorted_list)
  
  # fill in the condition for the while loop
  while left_pointer < right_pointer:
    # calculate the middle index using the two pointers
    mid_idx = (left_pointer + right_pointer) // 2  # floor division
    mid_val = sorted_list[mid_idx]
    if mid_val == target:
      return target
    if target < mid_val:
      # set the right_pointer to the appropriate value
      right_pointer = mid_idx
    if target > mid_val:
      # set the left_pointer to the appropriate value
      left_pointer = mid_idx + 1
  
  return "Value not in list" 

def merge_sort(items):
  if len(items) <= 1:
    return items

  middle_index = len(items) // 2
  left_split = items[:middle_index]
  right_split = items[middle_index:]

  left_sorted = merge_sort(left_split)
  right_sorted = merge_sort(right_split)

  return merge(left_sorted, right_sorted)

def merge(left, right):
  result = []

  while (left and right):
    if left[0] < right[0]:
      result.append(left[0])
      left.pop(0)
    else:
      result.append(right[0])
      right.pop(0)

  if left:
    result += left
  if right:
    result += right

  return result

array = randomArray(10)
print(array)
print()
sorted_array = bubble_sort_optimized(array)
print(sorted_array)
print()
random_number = array[random.randint(0,len(array) - 1)]
print(str(random_number) + " was found at index value " + str(linear_search(array, random_number)))
print()
print(str(binary_search(sorted_array, random_number)) + " was found") 
print()
array2 = randomArray(10000)
print(len(array2))
print()
time_start = time.time()
bubble_sort_optimized(array2)
time_end = time.time()
total_time = time_end - time_start
print("Optimized bubble sort took " + str(total_time) + " seconds")
print()
array3 = randomArray(10000)
time_start2 = time.time()
bubble_sort_unoptimized(array3)
time_end2 = time.time()
total_time2 = time_end2 - time_start2
print("Unoptimized bubble sort took " + str(total_time2) + " seconds")
time_start3 = time.time()
bubble_sort_optimized(array3)
time_end3 = time.time()
total_time3 = time_end3 - time_start3
print("Optimized bubble sort took " + str(total_time3) + " seconds")
print()
array4 = randomArray(10000)
time_start4 = time.time()
sorted_array2 = merge_sort(array4)
time_end4 = time.time()
total_time4 = time_end4 - time_start4
print("Merge sort took " + str(total_time4) + " seconds")
time_start5 = time.time()
bubble_sort_unoptimized(array4)
time_end5 = time.time()
total_time5 = time_end5 - time_start5
print("Unoptimized bubble sort took " + str(total_time5) + " seconds")
time_start6 = time.time()
bubble_sort_optimized(array4)
time_end6 = time.time()
total_time6 = time_end6 - time_start6
print("Optimized bubble sort took " + str(total_time6) + " seconds")
print()
sorted_array2[9999]=999
time_start7 = time.time()
print(str(binary_search(sorted_array2, 999)) + " was found")
time_end7 = time.time()
total_time7 = time_end7 - time_start7
print("Binary search took " + str(total_time7) + " seconds")
time_start8 = time.time()
print("999 was found at index value " + str(linear_search(sorted_array2, 999)))
time_end8 = time.time()
total_time8 = time_end8 - time_start8
print("Linear search took " + str(total_time8) + " seconds")
````
